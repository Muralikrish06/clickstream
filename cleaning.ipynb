```python
#MONTH MAPPER.
import pandas as pd
class MonthMapper:
    def __init__(self):
        self.month_map = {
            1: 'January',
            2: 'February',
            3: 'March',
            4: 'April',
            5: 'May',
            6: 'June',
            7: 'July',
            8: 'August',
            9: 'September',
            10: 'October',
            11: 'November',
            12: 'December'
        }
    def map_month(self, df):
        df['month'] = df['month'].map(self.month_map)
        return df

month_mapper = MonthMapper()
train = month_mapper.map_month(train)
display(train)  
```

```python
#COUNTRY MAPPING.
import pandas as pd

class CountryMapper:
    def __init__(self):
        self.country_map = {
            "1": "Australia", "2": "Austria", "3": "Belgium", "4": "British Virgin Islands",
            "5": "Cayman Islands", "6": "Christmas Island", "7": "Croatia", "8": "Cyprus",
            "9": "Czech Republic", "10": "Denmark", "11": "Estonia", "12": "unidentified",
            "13": "Faroe Islands", "14": "Finland", "15": "France", "16": "Germany", "17": "Greece",
            "18": "Hungary", "19": "Iceland", "20": "India", "21": "Ireland", "22": "Italy",
            "23": "Latvia", "24": "Lithuania", "25": "Luxembourg", "26": "Mexico",
            "27": "Netherlands", "28": "Norway", "29": "Poland", "30": "Portugal", "31": "Romania",
            "32": "Russia", "33": "San Marion", "34": "Slovakia", "35": "Slovenia", "36": "Spain",
            "37": "Sweden", "38": "Switzerland", "39": "Ukraine", "40": "United Arab Emirates",
            "41": "United Kingdom", "42": "USA", "43": "biz (.biz)", "44": "com (.com)",
            "45": "int (.int)", "46": "net (.net)", "47": "org (*.org)"
        }

    def map_countries(self, df):
        df = df.copy()
        df['country'] = df['country'].astype(str)
        df['country'] = df['country'].replace(self.country_map)
        df['country'] = df['country'].replace('nan', 'unidentified')
        return df


# Initialize the mapper
mapper = CountryMapper()

# Map countries
train = mapper.map_countries(train)
#test = mapper.map_countries(test)

# Display the results
display(train)
```

```python
import pandas as pd

class LabelEncoder:
    def __init__(self):
        self.label_map = {}
        self.inverse_label_map = {}

    def fit(self, df, column):
        # Generate label mapping for a column
        unique_values = df[column].unique()
        for index, value in enumerate(unique_values):
            self.label_map[value] = index
            self.inverse_label_map[index] = value

    def transform(self, df, column):
        # Transform the column using the label_map
        df[column] = df[column].map(self.label_map)
        return df

    def inverse_transform(self, df, column):
        # Inverse transform the column using inverse_label_map
        df[column] = df[column].map(self.inverse_label_map)
        return df
    
# Initialize the label encoder
label_encoder = LabelEncoder()

# Fit the encoder on the 'train' dataset, only on training columns
label_encoder.fit(train, 'country')
label_encoder.fit(train, 'page1_main_category')
label_encoder.fit(train, 'colour')
label_encoder.fit(train, 'location')
label_encoder.fit(train, 'model_photography')
label_encoder.fit(train, 'month')
label_encoder.fit(train, 'page2_clothing_model')

# Transform the 'train' dataset using the fitted label maps
train = label_encoder.transform(train, 'country')
train = label_encoder.transform(train, 'page1_main_category')
train = label_encoder.transform(train, 'colour')
train = label_encoder.transform(train, 'location')
train = label_encoder.transform(train, 'model_photography')
train = label_encoder.transform(train, 'month')
train = label_encoder.transform(train, 'page2_clothing_model')

# Save the encoded train data to CSV
train.to_csv("C:/Users/murkr/Downloads/train_data_encoded.csv", index=False)

# Now load the encoded data and inspect it
train_encoded = pd.read_csv("C:/Users/murkr/Downloads/train_data_encoded.csv")
display(train_encoded)
```

```python
import pandas as pd

class DataPreprocessor:
    def __init__(self):
        # Assuming that the mappers (MonthMapper, CountryMapper, etc.) are correctly implemented
        self.month_mapper = MonthMapper()
        self.country_mapper = CountryMapper()
        self.page1_mapper = Page1Mapper()
        self.colour_mapper = ColourMapper()
        self.location_mapper = LocationMapper()
        self.model_photo_mapper = ModelPhotoMapper()
        self.label_encoder = LabelEncoder()

    def preprocess(self, df, is_train=False):
        df = df.copy()
        
        # Apply mappers (replace this with your actual mapping classes)
        df = self.month_mapper.map_month(df)
        df = self.country_mapper.map_countries(df)
        df = self.page1_mapper.map_page1(df)
        df = self.colour_mapper.map_colours(df)
        df = self.location_mapper.map_locations(df)
        df = self.model_photo_mapper.map_model_photo(df)
        
        # Fit the label encoder only on the training data
        if is_train:
            self.label_encoder.fit(df, 'country')
            self.label_encoder.fit(df, 'page1_main_category')
            self.label_encoder.fit(df, 'colour')
            self.label_encoder.fit(df, 'location')
            self.label_encoder.fit(df, 'model_photography')
            self.label_encoder.fit(df, 'month')
            self.label_encoder.fit(df, 'page2_clothing_model')
        
        # Apply the label encoder transformation to both train and test data
        df = self.label_encoder.transform(df, 'country')
        df = self.label_encoder.transform(df, 'page1_main_category')
        df = self.label_encoder.transform(df, 'colour')
        df = self.label_encoder.transform(df, 'location')
        df = self.label_encoder.transform(df, 'model_photography')
        df = self.label_encoder.transform(df, 'month')
        df = self.label_encoder.transform(df, 'page2_clothing_model')

        return df

# Initialize the preprocessor
preprocessor = DataPreprocessor()

# For training data, we fit and transform
train = pd.read_csv("C:/Users/murkr/Downloads/train_data.csv")
train = preprocessor.preprocess(train, is_train=True)

# For test data, we only transform (using the encoder fitted on the train data)
test = pd.read_csv("C:/Users/murkr/Downloads/test_data.csv")
test_encoded = preprocessor.preprocess(test, is_train=False)

# Display the preprocessed test data
display(test_encoded)

# Save the preprocessed data to CSV
train.to_csv("C:/Users/murkr/Downloads/train_data_preprocessed.csv", index=False)
test_encoded.to_csv("C:/Users/murkr/Downloads/test_data_preprocessed.csv", index=False)
```
