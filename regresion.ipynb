```python
#FEATURE SELECTION FOR REGRESSION.
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, f_regression
train_cleaned = pd.read_csv("C:/Users/murkr/Downloads/train_data_cleaned.csv")
test_cleaned = pd.read_csv("C:/Users/murkr/Downloads/test_data_cleaned.csv")
display(train_cleaned.head())
display(test_cleaned)
display(train_cleaned.shape)
display(test_cleaned.shape)
```

```python
# Define the features and target
X_train=train_cleaned[['month','country','page1_main_category','colour','location','model_photography','page2_clothing_model',"page",'price_2']]
y_train=train_cleaned['price']

X_test=test_cleaned[['month','country','page1_main_category','colour','location','model_photography','page2_clothing_model','page','price_2']]
y_test=test_cleaned['price']
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
# Split training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Display data shapes
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_val shape:", X_val.shape)
print("y_val shape:", y_val.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python




#LINEAR REGRESSION.
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error,root_mean_squared_error

import mlflow
#import mlflow.sklearn
import os
# Start MLflow Experiment
mlflow.set_experiment("Linear_Regression")

with mlflow.start_run():
    # Train Model
    model_lr = LinearRegression()
    model_lr.fit(X_train, y_train)

    # Make Predictions
    y_pred_val = model_lr.predict(X_val)
    y_pred_test = model_lr.predict(X_test)

    # Calculate Validation Metrics
    mse_val_lr = mean_squared_error(y_val, y_pred_val)
    rmse_val_lr = np.sqrt(mse_val_lr)
    r2_val_lr = model_lr.score(X_val, y_val)

    # Log Validation Metrics to MLflow
    mlflow.sklearn.log_model(model_lr, "linear_regression_model")  # Naming the model
    mlflow.log_metric("MSE_Validation", mse_val_lr)
    mlflow.log_metric("RMSE_Validation", rmse_val_lr)
    mlflow.log_metric("R2_Validation", r2_val_lr)

    print(f"Validation MSE: {mse_val_lr}")
    print(f"Validation RMSE: {rmse_val_lr}")
    print(f"Validation R2: {r2_val_lr}")

    # Calculate Test Metrics
    mse_test_lr = mean_squared_error(y_test, y_pred_test)
    rmse_test_lr = np.sqrt(mse_test_lr)
    r2_test_lr = model_lr.score(X_test, y_test)

    # Log Test Metrics to MLflow
    mlflow.log_metric("MSE_Test", mse_test_lr)
    mlflow.log_metric("RMSE_Test", rmse_test_lr)
    mlflow.log_metric("R2_Test", r2_test_lr)

    print(f"Test MSE: {mse_test_lr}")
    print(f"Test RMSE: {rmse_test_lr}")
    print(f"Test R2: {r2_test_lr}")
```

```python
#RIDGE REGRESSION.
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error,root_mean_squared_error

import mlflow
import mlflow.sklearn
import os
# Start MLflow Experiment
mlflow.set_experiment("Ridge_Regression")

with mlflow.start_run():
    # Train Model
    model_ridge = Ridge()
    model_ridge.fit(X_train, y_train)

    # Make Predictions
    y_pred_val = model_ridge.predict(X_val)
    y_pred_test = model_ridge.predict(X_test)

    # Calculate Validation Metrics
    mse_val_ridge = mean_squared_error(y_val, y_pred_val)
    rmse_val_ridge = np.sqrt(mse_val_ridge)
    r2_val_ridge = model_ridge.score(X_val, y_val)

    # Log Validation Metrics to MLflow
    mlflow.sklearn.log_model(model_ridge, "ridge_regression_model")  # Naming the model
    mlflow.log_metric("MSE_Validation", mse_val_ridge)
    mlflow.log_metric("RMSE_Validation", rmse_val_ridge)
    mlflow.log_metric("R2_Validation", r2_val_ridge)

    print(f"Validation MSE: {mse_val_ridge}")
    print(f"Validation RMSE: {rmse_val_ridge}")
    print(f"Validation R2: {r2_val_ridge}")

    # Calculate Test Metrics
    mse_test_ridge = mean_squared_error(y_test, y_pred_test)
    rmse_test_ridge = np.sqrt(mse_test_ridge)
    r2_test_ridge = model_ridge.score(X_test, y_test)

    # Log Test Metrics to MLflow
    mlflow.log_metric("MSE_Test", mse_test_ridge)
    mlflow.log_metric("RMSE_Test", rmse_test_ridge)
    mlflow.log_metric("R2_Test", r2_test_ridge)

    print(f"Test MSE: {mse_test_ridge}")
    print(f"Test RMSE: {rmse_test_ridge}")
    print(f"Test R2: {r2_test_ridge}")
```

```python
#LASSO REGRESSION.
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error,root_mean_squared_error

import mlflow
import mlflow.sklearn
import os
# Start MLflow Experiment
mlflow.set_experiment("Lasso_Regression")

with mlflow.start_run():
    # Train Model
    model_lasso = Lasso()
    model_lasso.fit(X_train, y_train)

    # Make Predictions
    y_pred_val = model_lasso.predict(X_val)
    y_pred_test = model_lasso.predict(X_test)

    # Calculate Validation Metrics
    mse_val_lasso = mean_squared_error(y_val, y_pred_val)
    rmse_val_lasso = np.sqrt(mse_val_lasso)
    r2_val_lasso = model_lasso.score(X_val, y_val)

    # Log Validation Metrics to MLflow
    mlflow.sklearn.log_model(model_lasso, "lasso_regression_model")  # Naming the model
    mlflow.log_metric("MSE_Validation", mse_val_lasso)
    mlflow.log_metric("RMSE_Validation", rmse_val_lasso)
    mlflow.log_metric("R2_Validation", r2_val_lasso)

    print(f"Validation MSE: {mse_val_lasso}")
    print(f"Validation RMSE: {rmse_val_lasso}")
    print(f"Validation R2: {r2_val_lasso}")


    # Calculate Test Metrics
    mse_test_lasso = mean_squared_error(y_test, y_pred_test)
    rmse_test_lasso = np.sqrt(mse_test_lasso)
    r2_test_lasso = model_lasso.score(X_test, y_test)

    # Log Test Metrics to MLflow
    mlflow.log_metric("MSE_Test", mse_test_lasso)
    mlflow.log_metric("RMSE_Test", rmse_test_lasso)
    mlflow.log_metric("R2_Test", r2_test_lasso)

    print(f"Test MSE: {mse_test_lasso}")
    print(f"Test RMSE: {rmse_test_lasso }")
    print(f"Test R2: {r2_test_lasso}")
```

```python
#GRADIENT BOOSTING REGRESSION.
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error,root_mean_squared_error

import mlflow

import mlflow.sklearn
import os

# Start MLflow Experiment
mlflow.set_experiment("Gradient_Boosting_Regression")

with mlflow.start_run():
    # Train Model
    model_gb = GradientBoostingRegressor()
    model_gb.fit(X_train, y_train)

    # Make Predictions
    y_pred_val = model_gb.predict(X_val)
    y_pred_test = model_gb.predict(X_test)

    # Calculate Validation Metrics
    mse_val_gb = mean_squared_error(y_val, y_pred_val)
    rmse_val_gb = np.sqrt(mse_val_gb)
    r2_val_gb = model_gb.score(X_val, y_val)

    # Log Validation Metrics to MLflow
    mlflow.sklearn.log_model(model_gb, "gradient_boosting_regression_model")  # Naming the model
    mlflow.log_metric("MSE_Validation", mse_val_gb)
    mlflow.log_metric("RMSE_Validation", rmse_val_gb)
    mlflow.log_metric("R2_Validation", r2_val_gb)

    print(f"Validation MSE: {mse_val_gb}")
    print(f"Validation RMSE: {rmse_val_gb}")
    print(f"Validation R2: {r2_val_gb}")

    # Calculate Test Metrics
    mse_test_gb = mean_squared_error(y_test, y_pred_test)
    rmse_test_gb = np.sqrt(mse_test_gb)
    r2_test_gb = model_gb.score(X_test, y_test)

    # Log Test Metrics to MLflow
    mlflow.log_metric("MSE_Test", mse_test_gb)
    mlflow.log_metric("RMSE_Test", rmse_test_gb)
    mlflow.log_metric("R2_Test", r2_test_gb)


    print(f"Test MSE: {mse_test_gb}")
    print(f"Test RMSE: {rmse_test_gb}")
    print(f"Test R2: {r2_test_gb}")
```
